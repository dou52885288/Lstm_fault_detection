```python
import pandas as pd 
icing_features = pd.read_csv('E:/PycharmProjects/AI/icing_residuals.csv')
icing_features
```

```python
import numpy as np 
n = len(icing_features)
tm, tp, tpp, te, tr = np.zeros(n),np.zeros(n),np.zeros(n),np.zeros(n),np.zeros(n)
for i in range(n):
    data = np.array(icing_features['yaw_position'][i:i+10])
    tm[i] = np.sum(data)/10
    tp[i] = np.max(data)
    tpp[i] = np.subtract(np.max(data),np.min(data))
    te[i] = np.sum(np.square(data))
    tr[i] = (np.sum(np.sqrt(np.abs(data)))/10)**2
tm
```

```python
f1 = np.vstack((tm,tp,tpp,te,tr)).T
f1.shape
```
