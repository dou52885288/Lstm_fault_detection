```python
import pandas as pd 
cha_b1m1 = pd.read_csv('E:/PycharmProjects/AI/cc_b1m1.csv')
cha_b2m1 = pd.read_csv('E:/PycharmProjects/AI/cc_b2m1.csv')
cha_b3m1 = pd.read_csv('E:/PycharmProjects/AI/cc_b3m1.csv')
cha = pd.concat([cha_b1m1.iloc[:,1],cha_b2m1.iloc[:,1],cha_b3m1.iloc[:,1]],axis=1)
cha
```

```python
import numpy as np 
tm, tp, tpp, te, tr = np.zeros(len(cha)),np.zeros(len(cha)),np.zeros(len(cha)),np.zeros(len(cha)),np.zeros(len(cha))
for i in range(len(cha)):
    data = np.array(cha['cc_b1m1'][i:i+10])
    tm[i] = np.sum(data)/10
    tp[i] = np.max(data)
    tpp[i] = np.subtract(np.max(data),np.min(data))
    te[i] = np.sum(np.square(data))
    tr[i] = (np.sum(np.sqrt(np.abs(data)))/10)**2
tm
```

```python
b1_fea = np.vstack((tm,tp,tpp,te,tr)).T
b1_fea.shape
```

```python
tm1, tp1, tpp1, te1, tr1 = np.zeros(len(cha)),np.zeros(len(cha)),np.zeros(len(cha)),np.zeros(len(cha)),np.zeros(len(cha))
for i in range(len(cha)):
    data = np.array(cha['cc_b2m1'][i:i+10])
    tm1[i] = np.sum(data)/10
    tp1[i] = np.max(data)
    tpp1[i] = np.subtract(np.max(data),np.min(data))
    te1[i] = np.sum(np.square(data))
    tr1[i] = (np.sum(np.sqrt(np.abs(data)))/10)**2
b2_fea = np.vstack((tm1,tp1,tpp1,te1,tr1)).T
b2_fea.shape
```

```python
tm2, tp2, tpp2, te2, tr2 = np.zeros(len(cha)),np.zeros(len(cha)),np.zeros(len(cha)),np.zeros(len(cha)),np.zeros(len(cha))
for i in range(len(cha)):
    data = np.array(cha['cc_b3m1'][i:i+10])
    tm2[i] = np.sum(data)/10
    tp2[i] = np.max(data)
    tpp2[i] = np.subtract(np.max(data),np.min(data))
    te2[i] = np.sum(np.square(data))
    tr2[i] = (np.sum(np.sqrt(np.abs(data)))/10)**2
b3_fea = np.vstack((tm2,tp2,tpp2,te2,tr2)).T
b3_fea.shape
```

```python
features = np.concatenate((b1_fea,b2_fea,b3_fea),axis=1)
features.shape
```

```python
features = pd.DataFrame(features,columns=['b1_tm','b1_tp','b1_tpp','b1_te','b1_tr','b2_tm','b2_tp','b2_tpp','b2_te','b2_tr',\
                                  'b3_tm','b3_tp','b3_tpp','b3_te','b3_tr'])
features
```

```python
features.to_csv('E:/PycharmProjects/AI/features1.csv')
```
