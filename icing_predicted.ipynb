```python
import pandas as pd 
import xlrd

#打开索引表
excel_name = 'E:/PycharmProjects/AI/15/xulie.xlsx' 
wb = xlrd.open_workbook(excel_name)

# 获取workbook中所有的表格
sheets = wb.sheet_names()

# 循环遍历所有sheet，第一个sheet是故障时间索引，第二个是正常状态索引
df = pd.DataFrame()
for i in range(len(sheets)):
    df1 = pd.read_excel(excel_name, sheet_name=i)
    df = pd.concat([df,df1],ignore_index=True)
df
```

```python
data = pd.read_csv('E:/PycharmProjects/AI/15/15_data.csv')
data.head()
```

```python
#数据太长，截取其中几段组成训练集和测试集
failure = pd.DataFrame()
for i in range(29):
    failure = pd.concat([failure,data.iloc[df['start point'][i]-1:df['end point'][i],1:27]],axis=0,ignore_index=True)
failure.iloc[:,2].plot()
```

```python
normal = pd.DataFrame()
for i in range(6):
    normal = pd.concat([normal,data.iloc[df['start point'][i+29]-1:df['end point'][i+29],1:27]],ignore_index=True)
normal.iloc[:,2].plot()
```

```python
x_train = normal.iloc[:80000,:]
x_test = pd.concat([normal.iloc[80000:,:],failure],ignore_index=True)
print(x_train.shape,x_test.shape)
```

```python
from scipy import signal 
Fs = 100
wp = 5/(Fs/2)
ws = 10/(Fs/2)
rp = 2
rs = 40
N, Wn = signal.buttord(wp, ws, rp, rs)
b, a = signal.butter(N, Wn,'lowpass')
for i in range(26):
    x_train.iloc[:,i] = signal.filtfilt(b,a,x_train.iloc[:,i])
    x_test.iloc[:,i] = signal.filtfilt(b,a,x_test.iloc[:,i])
x_train.head()
```

```python
x_test.shape
```

```python
xtrain =[]
ytrain =[]
xtest  =[]
ytest =[]
for i in range(15, len(x_train)):
    xtrain.append(x_train.iloc[i - 15:i, :].values)
    ytrain.append(x_train.iloc[i, 8])
for j in range(15, len(x_test)):
    xtest.append(x_test.iloc[j - 15:j, :].values)
    ytest.append(x_test.iloc[j, 8])
print(len(xtrain),len(xtest))
```

```python
import tensorflow as tf
import numpy as np
np.random.seed(7)
np.random.shuffle(xtrain)
np.random.seed(7)
np.random.shuffle(ytrain)
tf.random.set_seed(7)
# 将训练集由list格式变为array格式
xtrain, ytrain = np.array(xtrain), np.array(ytrain)
xtest, ytest = np.array(xtest), np.array(ytest)

# 使x_train符合RNN输入要求：[送入样本数， 循环核时间展开步数， 每个时间步输入特征个数]。
xtrain = np.reshape(xtrain, (xtrain.shape[0], 15, 26))
xtest = np.reshape(xtest, (xtest.shape[0], 15, 26))
from tensorflow.keras.layers import LSTM, Dense, Dropout
model = tf.keras.Sequential([
    LSTM(80, return_sequences=True),
    Dropout(0.2),
    LSTM(100),
    Dropout(0.2),
    Dense(1)
])

model.compile(optimizer=tf.keras.optimizers.Adam(0.001),
              loss='mean_squared_error')  # 损失函数用均方误差


history = model.fit(xtrain, ytrain, batch_size=128, epochs=5, validation_data=(xtest, ytest), validation_freq=1)

model.summary()
```

```python
loss = history.history['loss']
val_loss = history.history['val_loss']
import matplotlib.pyplot as plt
plt.plot(loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()
```

```python
predicted_value = model.predict(xtest)
predicted = np.reshape(predicted_value,(predicted_value.shape[0],))
real = ytest
#画出真实数据和预测数据的对比曲线
plt.figure()
plt.plot(real, color='red', label='real value')
plt.plot(predicted, color='blue', label='Predicted value')
plt.title('pitch1 Prediction')
plt.xlabel('Time')
plt.ylabel('pitch1')
plt.legend()
plt.show()
```
